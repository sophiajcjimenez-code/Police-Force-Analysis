```python
# first merging crimedf with outcomedf to isolate problems if any
merged_df = pd.merge(
    crimedf,
    outcomedf,
    on="Crime ID",
    how="left",
    suffixes=("_crime", "_outcome")
)
```

```python
# removing duplicate columns
merged_df = merged_df.drop(columns=[
    "Longitude_outcome",
    "Latitude_outcome",
    "Year_outcome",
    "Month_outcome",
    "LSOA code_outcome",
    "LSOA name_outcome",
    "Location_outcome",
    "Reported by_outcome",
    "Falls within_outcome",
    "District_outcome"
])
```

```python
# merged_df["District_crime"] = merged_df["District_crime"].fillna("Unknown")
merged_df["Outcome type"] = merged_df["Outcome type"].fillna("Unknown")
```

```python
# merging sub df with populationdf to make main df
df = pd.merge(
    merged_df,
    populationdf,
    left_on="LSOA code_crime",
    right_on="LSOA 2021 Code",
    how="inner"   # keeps records that only matches
)
```

```python
# changing the names of column titles

df = df.rename(columns={
    "Year_crime": "Year",
    "Month_crime": "Month",
    "Reported by_crime" : "Reported by",
    "Longitude_crime": "Longitude",
    "Latitude_crime": "Latitude",
    "Location_crime": "Location",
    "LSOA code_crime": "LSOA code",
    "LSOA name_crime": "LSOA name",
    "District_crime": "District",
    "Total" : "Population"
})
```
